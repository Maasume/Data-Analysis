# Introduction
In this Project, the pandas library is used to analyse the data. The data is related to Austin, and it is imported by API. Then, using the head method, the first 15 rows and the column names are also demonstrated. In this report, the limit is considered to be around 3,000,000.

```python
import pandas as pd
from sodapy import Socrata

client = Socrata("data.austintexas.gov", "vT9dyEXWAk1UrzgBe8KXFuaU9")


results = client.get("7d8e-dm7r", limit=3000000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

# cleaning data
Before answering any questions, some preliminary data cleaning may be required, such as removing empty rows, rows with "None" and "Nan", dropping unnecessary columns in a DataFrameor, rows lacking critical data.

```python
# cleaning data
results_df.dropna(axis='rows')

def get_empty_rows(results_df):
    indices = []
    for i, s in results_df.iterrows():
        for v in s.values:
            if str(v).lower() == 'nan':
                if i not in indices:
                    indices.append(i)

    return indices

def get_empty_rows2(results_df):
    indices = []
    for i, s in results_df.iterrows():
        for v in s.values:
            if str(v).lower() == 'null':
                if i not in indices:
                    indices.append(i)

    return indices

list_of_empty_rows = get_empty_rows(results_df)
list_of_empty_rows2 = get_empty_rows(results_df)

clean_data = results_df.drop(list_of_empty_rows+list_of_empty_rows2)
```

# Q1: Compute then the O-D matrix, i.e., the number of bookings starting in council district i and ending in council district j. Try to visualize the results in a meaningful way.

Before computing the O-D matrix , the value of needed columns should be replace bad values and convert them to integers. Then, the pivot table is calculated. The table demonstrates that the busiest area is "district 9" and the least traffic is belong to "district 6". Furthermore, the "district 9" has the highest number of travels that starts and ends in it.

```python
# To convert datatypes and to replace some bad values
clean_data["council_district_start"] = clean_data["council_district_start"].replace("None", 0)
clean_data["council_district_start"] = clean_data["council_district_start"].astype('int')
clean_data["council_district_end"] = clean_data["council_district_end"].replace("None", 0)
clean_data["council_district_end"] = clean_data["council_district_end"].astype('int')

# OD Matrix calculation using pivot table
matrix = (
    clean_data.assign(count=1)
    .pivot_table(index="council_district_start", columns="council_district_end",
                 values="count", aggfunc="count")
    .fillna(0)
    .astype(int)
).sort_values("council_district_start")

matrix
```
