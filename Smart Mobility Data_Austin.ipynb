```python
import pandas as pd
from sodapy import Socrata

client = Socrata("data.austintexas.gov", "vT9dyEXWAk1UrzgBe8KXFuaU9")

results = client.get("7d8e-dm7r", limit=3000000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

```python
results_df.head(15)
```

# cleaning data
Before answering any questions, some preliminary data cleaning may be required, such as removing empty rows, rows with "None" and "Nan", dropping unnecessary columns in a DataFrameor, rows lacking critical data.

```python
# cleaning data
results_df.dropna(axis='rows')

def get_empty_rows(results_df):
    indices = []
    for i, s in results_df.iterrows():
        for v in s.values:
            if str(v).lower() == 'nan':
                if i not in indices:
                    indices.append(i)

    return indices

def get_empty_rows2(results_df):
    indices = []
    for i, s in results_df.iterrows():
        for v in s.values:
            if str(v).lower() == 'null':
                if i not in indices:
                    indices.append(i)

    return indices

list_of_empty_rows = get_empty_rows(results_df)
list_of_empty_rows2 = get_empty_rows(results_df)

clean_data = results_df.drop(list_of_empty_rows+list_of_empty_rows2)
```
