# Introduction
In this project, the pandas library is used to analyse the data. The data is related to Austin, and it is imported by API. Then, using the head method, the first 15 rows and the column names are also demonstrated. In this report, the limit is considered to be around 3,000,000.

```python
import pandas as pd
from sodapy import Socrata

client = Socrata("data.austintexas.gov", "vT9dyEXWAk1UrzgBe8KXFuaU9")

results = client.get("7d8e-dm7r", limit=3000000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

```python
results_df.head(15)
```

# cleaning data
Before answering any questions, some preliminary data cleaning may be required, such as removing empty rows, rows with "None" and "Nan", dropping unnecessary columns in a DataFrameor, rows lacking critical data.

```python
# cleaning data
results_df.dropna(axis='rows')

def get_empty_rows(results_df):
    indices = []
    for i, s in results_df.iterrows():
        for v in s.values:
            if str(v).lower() == 'nan':
                if i not in indices:
                    indices.append(i)

    return indices

def get_empty_rows2(results_df):
    indices = []
    for i, s in results_df.iterrows():
        for v in s.values:
            if str(v).lower() == 'null':
                if i not in indices:
                    indices.append(i)

    return indices

list_of_empty_rows = get_empty_rows(results_df)
list_of_empty_rows2 = get_empty_rows(results_df)

clean_data = results_df.drop(list_of_empty_rows+list_of_empty_rows2)
```

## The mean and median of utilization percentages of vehicles during different days of the week

```python
# To convert datatypes to integer
clean_data["trip_duration"] = clean_data["trip_duration"].astype('int')

mean_days = clean_data.groupby('day_of_week', as_index=False)["trip_duration"].mean()

mean_days
```

```python
mean_days_pt = mean_days.plot(kind="bar",x='day_of_week', color='purple')
```

```python
median_days = clean_data.groupby("day_of_week", as_index=False)["trip_duration"].median()
median_days
```

```python
median_days_pt = mean_days.plot(kind="bar",x='day_of_week', color='red')
```
